```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/GOOG?period1=1579365747&period2=1610988147&interval=1d&events=history&includeAdjustedClose=true')
```

```
df
```

```
x = df['Close']   #  x object, which contains the daily close price of the stock.
```

```
returns = x.pct_change(1).dropna()
```

```
import matplotlib.pyplot as plt
```

```
plt.hist(returns,bins="rice",label="Daily close price") # histogram of returns.
plt.legend()
plt.show()
```

```
plt.boxplot(returns,labels=["Daily close price"])  #  a boxplot
plt.show()
```

```
plt.plot(returns) # Volatility
plt.xlabel("Time")
plt.ylabel("Daily returns")
plt.show()
```

```
plt.plot(returns.rolling(20).std()) 
plt.xlabel("Time")
plt.ylabel("20-days rolling standard deviation")
plt.show()
```

```
from statsmodels.graphics.tsaplots import plot_pacf
```

```
plot_pacf(returns,lags=20)
```
